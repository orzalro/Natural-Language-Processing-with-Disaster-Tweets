```python
import pandas as pd

train_data = pd.read_csv('/kaggle/input/nlp-getting-started/train.csv')
print('훈련 데이터 총 갯수 :', len(train_data))
train_data.head()
```

```python
print(train_data['keyword'].value_counts(normalize=True, dropna=False))
print(train_data['location'].value_counts(normalize=True, dropna=False))
```

```python
crosstab = pd.crosstab(train_data['keyword'], train_data['target'], normalize='index')
print(crosstab[1].sort_values(ascending=False)) # 키워드별 재앙일 확률
```

```python
train_data['text_length'] = train_data['text'].apply(len)
print(train_data['text_length'].describe())

import matplotlib.pyplot as plt

plt.hist(train_data['text_length'], bins=50)
plt.xlabel('Text Length')
plt.ylabel('Count')
plt.title('Distribution of Text Lengths')
plt.show()
```
